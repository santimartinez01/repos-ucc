```python
# Cargar los datos
data = pd.read_csv('chicago_crimes.csv')
```

```python
# Mostrar las primeras filas de los datos
print(data.head())
```

```python
# Descripción estadística de los datos numéricos
print(data.describe())
```

```python
# Distribución de características categóricas
categorical_columns = ['Primary Type', 'Description', 'Location Description']
for col in categorical_columns:
 print(f"Distribución de {col}:")
 print(data[col].value_counts())
```

```python
# Ajusta el estilo y paleta
sns.set(style="whitegrid")
sns.set_palette("pastel")

# Genera los histogramas
data.hist(figsize=(12, 10), bins=30, edgecolor='black', grid=False)

# Ajusta la estética general
plt.tight_layout()

# Muestra los gráficos
plt.show()
```

```python
# Convertir datos no numericos en numericos
for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Configurar estilo de Seaborn
sns.set(style="whitegrid", palette="pastel", font_scale=1.2)

# Crear el mapa de calor
corr_matrix = data.corr()
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'Correlación'})

# Títulos y etiquetas
plt.title('Mapa de Calor de Correlación', fontsize=18)
plt.xlabel('Variables', fontsize=14)
plt.ylabel('Variables', fontsize=14)

# Mostrar gráfico
plt.show()
```
